```python
import pandas as pd
```

```python
df = pd.read_csv('reviews.csv')
```

```python
df.head()
```

```python
df.drop(['reviewId','userName','userImage','score','thumbsUpCount','reviewCreatedVersion','at','replyContent','repliedAt','sortOrder','appId'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns = {'content':'Review'}, inplace = True) 
```

```python
df["word_count"]=df['Review'].apply(lambda x:len(x.split()))
```

```python
df.head()
```

```python
df["char_count"]=df["Review"].apply(lambda x:len(x))
```

```python
def average_words(x):
    words=x.split()
    return sum(len(word)for word in words)/len(words)

df["average_word_length"]=df["Review"].apply(lambda x: average_words(x))
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop_words=stopwords.words('english')
```

```python
stop_words
```

```python
len(stop_words)
```

```python
df["stop_word_count"]=df["Review"].apply(lambda x:len([word for word in x.split() if word.lower() in stop_words]))
```

```python
df.head()
```

```python
df["stop_word_rate"]=df["stop_word_count"]/df["word_count"]
```

```python
df.sort_values(by="stop_word_rate")
```

```python
df.describe()
```

```python
df["lowercase"]=df["Review"].apply(lambda x:" ".join(word.lower()for word in x.split()))
```

```python
df.head()
```

```python
df["punctuation"]=df["lowercase"].str.replace('[^\w\s]','')
```

```python
df
```

```python
df["stopwords"]=df["punctuation"].apply(lambda x:" ".join(word for word in x.split()if word not in stop_words))
```

```python
df.head()
```

```python
pd.Series("".join(df["stopwords"]).split()).value_counts()[:30]
```

```python
other_stop_words=["app","tasks","like","use","time","would"]
```

```python
len(other_stop_words)
```

```python
df["clean_review"]=df["stopwords"].apply(lambda x:" ".join(word for word in x.split()if word not in other_stop_words))
```

```python
pd.Series("".join(df["clean_review"]).split()).value_counts()[:30]
```

```python
df.head()
```

```python
from textblob import Word
```

```python
import nltk  
nltk.download()
```

```python
df["lemmatized"]=df["clean_review"].apply(lambda x:" ".join(Word(word).lemmatize()for word in x.split()))
```

```python
df.head()
```

```python
from textblob import TextBlob
```

```python
df["lemmatized"].apply(lambda x:TextBlob(x).sentiment)
```

```python
df["polarity"]=df["lemmatized"].apply(lambda x:TextBlob(x).sentiment[0])
```

```python
df["subjectivity"]=df["lemmatized"].apply(lambda x:TextBlob(x).sentiment[1])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.sort_values(by="polarity")
```

```python
df.drop(["lowercase","punctuation","stopwords","clean_review"],axis=1,inplace=True)
```
